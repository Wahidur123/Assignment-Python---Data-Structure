Q5.Describe the key features of sets and provide examples of their use

```python
#Unordered:Sets do not maintain any order of elements. The elements in a set are stored in an arbitrary order, and you cannot index or slice them like you can with lists or tuples.
my_set = {10, 20, 30}
print(my_set)
```

Mutable

    Sets are mutable, meaning you can add and remove elements after the set is created. However, since sets do not allow duplicate elements, adding a duplicate value has no effect.

```python
my_set = {1, 2, 3}
my_set.add(4)
print(my_set)
```

```python
my_set = {1, 2, 3}
my_set.remove(2)
print(my_set)
```

```python
#Unique Elements:Sets do not allow duplicate elements. If you try to add a duplicate element, the set will automatically ignore it.
my_set = {1, 2, 2, 3, 4, 4}
print(my_set)
```

```python
#No Indexing, Slicing, or Concatenation:Sets do not support indexing, slicing, or concatenation, since they are unordered. This is a key difference compared to lists and tuples.
my_set = {1, 2, 2, 3, 4, 4}
my_set[0]
```

# Q6.discuss the use cases of tuples and sets in Python programming.

Both tuples and sets are important data structures in Python, and they are used in different scenarios depending on their properties—tuples are ordered and immutable, while sets are unordered and mutable but only contain unique elements.

```python
#TUPLES
#Storing Fixed Data (Immutability):Tuples are ideal for storing data that should not be modified after creation. Because tuples are immutable, once they are created, their values cannot be changed. This is useful for storing constant data or configuration settings that should not be altered during program execution.
config = ("localhost", 8080)
```

```python
#Using as Dictionary Keys:Tuples can be used as keys in dictionaries, as they are hashable (immutable). This is not possible with lists since they are mutable. A tuple is often used to represent a pair of coordinates or any other fixed combination of values.
location = {(40.7128, -74.0060): "New York", (34.0522, -118.2437): "Los Angeles"}
print(location[(40.7128, -74.0060)])
```

```python
#Memory Efficiency:Since tuples are immutable, they are generally more memory-efficient than lists for storing a fixed set of items. This is especially important when you need to store large amounts of data that should not change during the program’s execution.
# A tuple of values that shouldn't change
months = ("January", "February", "March", "April", "May", "June")
```

```python
#Sets
#Removing Duplicates:A common use case for sets is to eliminate duplicates from a collection. Since sets do not allow duplicate elements, converting a list or other iterable to a set removes any repeated items.
numbers = [1, 2, 2, 3, 4, 4, 5]
unique_numbers = set(numbers)
print(unique_numbers)
```

```python
#Efficient Membership Testing:Sets are particularly useful when you need to frequently check whether an element exists in a collection. Membership tests (in operator) in sets are on average O(1) in time complexity, which is much faster than testing membership in lists or tuples (which are O(n)).
allowed_users = {"Alice", "Bob", "Charlie"}
print("Alice" in allowed_users)  # Output: True
print("Eve" in allowed_users)    # Output: False
```

Q7.Describe how to add, modify, and delete items in a dictionary with examples.

In Python, a dictionary is an unordered, mutable collection of key-value pairs. You can add, modify, and delete items in a dictionary using various methods.

```python
#Adding Items to a Dictionary:To add a new item (key-value pair) to a dictionary, you can use the assignment operator (=) with the desired key.
# Initialize an empty dictionary
my_dict = {}

# Adding a new key-value pair
my_dict["name"] = "Alice"
my_dict["age"] = 30

print(my_dict)  # Output: {'name': 'Alice', 'age': 30}
```

```python
# Existing dictionary
person = {"name": "Alice", "age": 30}

# Adding a new key-value pair
person["city"] = "New York"

print(person)  # Output: {'name': 'Alice', 'age': 30, 'city': 'New York'}
```

```python
#Modifying Items in a Dictionary:To modify the value of an existing key in a dictionary, you can use the assignment operator (=) with the key and assign it a new value.
# Existing dictionary
person = {"name": "Alice", "age": 30}

# Modifying the value associated with the key "age"
person["age"] = 31

print(person)  # Output: {'name': 'Alice', 'age': 31}
```

```python
#Deleting Items from a Dictionary
# Existing dictionary
person = {"name": "Alice", "age": 31, "city": "New York"}

# Deleting the "city" key-value pair
del person["city"]

print(person)  # Output: {'name': 'Alice', 'age': 31}
```

 Q8.Discuss the importance of dictionary keys being immutable and provide examples.

In Python, dictionary keys must be immutable objects. This is a fundamental requirement that ensures the integrity of the dictionary’s internal structure and allows the dictionary to perform fast lookups

```python
#Mutability and Key Integrity:If keys were mutable, and one of them was altered after it was inserted into the dictionary, the dictionary would no longer be able to find it properly because the hash value would have changed. This could cause unpredictable behavior and bugs that are difficult to debug.
d = {}
key = [1, 2, 3]  # List is mutable
d[key] = "value"
print(d)
```

```python
#Immutability and Hashing in Practice:Since immutable objects are used as dictionary keys, their hash values are fixed and can be used reliably for quick lookups, insertions, and deletions. This is the core reason why dictionaries in Python perform so efficiently. For example, when you use a string as a key, Python uses the string’s hash value to find it in the dictionary:
d = {"apple": 1, "banana": 2}
print(d["apple"])  # Fast O(1) lookup using the hash of the key
```
